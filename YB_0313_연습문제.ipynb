# Occupation

### Introduction:

Special thanks to: https://github.com/justmarkham for sharing the dataset and materials.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|', index_col='user_id')
print(users.head())
```

### Step 4. Discover what is the mean age per occupation

```python
mean_age = users.groupby('occupation')['age'].mean()
print(mean_age)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
male_ratio = (users['gender'] == 'M').astype(int).groupby(users['occupation']).mean()
sorted_male_ratio = male_ratio.sort_values(ascending=False)
print(sorted_male_ratio)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_max_ages = users.groupby('occupation')['age'].agg(['min', 'max'])
print(min_max_ages)
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_comb = users.groupby(['occupation', 'gender'])['age'].mean()
print(mean_age_comb)
```

### Step 8.  For each occupation present the percentage of women and men

```python
gender_percentage = users.groupby('occupation')['gender'].value_counts(normalize=True)*100
print(gender_percentage)
```
